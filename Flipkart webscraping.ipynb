```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def extract_record(url):
    record_req = requests.get(url)
    soup1 = BeautifulSoup(record_req.text,'html.parser')
    
    contents = []
    
    
    try:
        a_string = soup1.find('span',attrs={'class':'B_NuCI'}).text
        Brand = a_string.split()[0].strip()
    except AttributeError:
        return
    
    try:
        discounted_price = soup1.find('div',attrs={'class':'_30jeq3 _16Jk6d'}).text
        discounted_price = discounted_price.replace("₹",'')
    except AttributeError:
        discounted_price = ''
        
    try:
        original_price = soup1.find('div',attrs={'class':'_3I9_wc _2p6lqe'}).text
        original_price = original_price.replace("₹",'')
    except AttributeError:
        original_price = discounted_price
        
    try:
        discount = soup1.find('div',attrs={'class':'_3Ay6Sb _31Dcoz'}).text
    except AttributeError:
        discount =''
     
    try:
        rating =   soup1.find('div',attrs={'class':'_3LWZlK'}).text   
    except AttributeError:
        rating = ''
       
    try:
        rating_count =   soup1.find('span',attrs={'class':'_2_R_DZ'}).text   
    except AttributeError:
        rating_count = ''



    table_count = soup1.find_all('table',attrs={'class':'_14cfVK'})
    for data_table in table_count:
        rows = data_table.find_all('tr')
        for row in rows:
            td = row.find_all('td')
            row1=[i.text.strip() for i in td]
            contents.append(row1)
        
    # Convert the Table contents to Data Frame    
    Mobile_data = pd.DataFrame(contents)

    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Model Name'])
        if boolean_finding:
            test = Mobile_data.loc[Mobile_data[0]== 'Model Name'].index[0]
            ModelName = Mobile_data.loc[test,1]
        else:
            ModelName = ''
            
    except AttributeError:
        ModelName = ''
#     print(ModelName)

    # Get Model Number
    try:
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Model Number'])
        if boolean_finding:
            test = Mobile_data.loc[Mobile_data[0]== 'Model Number'].index[0]
            ModelNumber = Mobile_data.loc[test,1]
        else:
            ModelNumber = ''
    except AttributeError:
        ModelNumber = ''
#     print(ModelNumber)

    # Get Sim type
    try:
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'SIM Type'])
        if boolean_finding:
            test = Mobile_data.loc[Mobile_data[0]== 'SIM Type'].index[0]
            Simtype = Mobile_data.loc[test,1]
        else:
            Simtype = ''

    except AttributeError:
        Simtype = ''
#     print(Simtype)

    # Get Technology type
    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Browse Type'])
        if boolean_finding:    
            test = Mobile_data.loc[Mobile_data[0]== 'Browse Type'].index[0]
            Technology = Mobile_data.loc[test,1]
        else:
            Technology = ''
    except AttributeError:
        Technology = ''
#     print(Technology)


    # Get Sim Size type
    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'SIM Size'])
        if boolean_finding:      
            test = Mobile_data.loc[Mobile_data[0]== 'SIM Size'].index[0]
            Simsize = Mobile_data.loc[test,1]
        else:
            Simsize = ''
    except AttributeError:
        Simsize = ''
#     print(Simsize)

    # Get Screen type
    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Touchscreen'])
        if boolean_finding:     
            test = Mobile_data.loc[Mobile_data[0]== 'Touchscreen'].index[0]
            Touchscreen = Mobile_data.loc[test,1]
        else:
            Touchscreen =''
                
    except AttributeError:
        Touchscreen = ''
#     print(Touchscreen)

    # Get Display Size 
    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Display Size'])
        if boolean_finding:      
            test = Mobile_data.loc[Mobile_data[0]== 'Display Size'].index[0]
            Displaysize = Mobile_data.loc[test,1]
        else:
            Displaysize = ''
    except AttributeError:
        Displaysize = ''
#     print(Displaysize)

    # Get Primary Camera Available 
    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Primary Camera Available'])
        if boolean_finding:     
            test = Mobile_data.loc[Mobile_data[0]== 'Primary Camera Available'].index[0]
            PCAvailable = Mobile_data.loc[test,1]
        else:
            PCAvailable = ''
    except AttributeError:
        PCAvailable = ''
#     print(PCAvailable)

    # Get Primary Camera info 
    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Primary Camera'])
        if boolean_finding:      
            test = Mobile_data.loc[Mobile_data[0]== 'Primary Camera'].index[0]
            Primarycamera = Mobile_data.loc[test,1]
        else:
            Primarycamera = ''
    except AttributeError:
        Primarycamera = ''
#     print(Primarycamera)


    # Get Secondary Camera Available 
    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Secondary Camera Available'])
        if boolean_finding:      
            test = Mobile_data.loc[Mobile_data[0]== 'Secondary Camera Available'].index[0]
            SCAvailable = Mobile_data.loc[test,1]
        else:
            SCAvailable = ''
    except AttributeError:
        SCAvailable = ''
#     print(SCAvailable)


    # Get Secondary Camera info 
    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Secondary Camera'])
        if boolean_finding:      
            test = Mobile_data.loc[Mobile_data[0]== 'Secondary Camera'].index[0]
            Secondarycamera = Mobile_data.loc[test,1]
        else:
            Secondarycamera = ''
    except AttributeError:
        Secondarycamera = ''
#     print(Secondarycamera)

    # Get Internal Storage 
    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Internal Storage'])
        if boolean_finding:      
            test = Mobile_data.loc[Mobile_data[0]== 'Internal Storage'].index[0]
            Intstorage = Mobile_data.loc[test,1]
        else:
            Intstorage = ''
    except AttributeError:
        Intstorage = ''
#     print(Intstorage)

    # Get RAM Info 
    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'RAM'])
        if boolean_finding:      
            test = Mobile_data.loc[Mobile_data[0]== 'RAM'].index[0]
            Ram = Mobile_data.loc[test,1]
        else:
            Ram = ''
    except AttributeError:
        Ram = ''
#     print(Ram)

    # Get Battery Info 
    try:
#         Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Battery Capacity'])
        if boolean_finding:  
            test = Mobile_data.loc[Mobile_data[0]== 'Battery Capacity'].index[0]
            Battery = Mobile_data.loc[test,1]
        else:
            Battery = ''
    except AttributeError:
        Battery = ''
#     print(Battery)

    # Get Expandable Storage Info 
    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Expandable Storage'])
        if boolean_finding:      
            test = Mobile_data.loc[Mobile_data[0]== 'Expandable Storage'].index[0]
            Exstorage = Mobile_data.loc[test,1]
        else:
            Exstorage = ''
    except AttributeError:
        Exstorage = ''
#     print(Exstorage)

    # Get Memory Card Type Info 
    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Supported Memory Card Type'])
        if boolean_finding:      
            test = Mobile_data.loc[Mobile_data[0]== 'Supported Memory Card Type'].index[0]
            Memcardtype = Mobile_data.loc[test,1]
        else:
            Memcardtype = ''
    except AttributeError:
        Memcardtype = ''
#     print(Memcardtype)


    # Get Memory Card slot Type Info 
    try:
    #     Mobile_data = pd.DataFrame(contents)
        boolean_finding = len(Mobile_data[Mobile_data[0] == 'Memory Card Slot Type'])
        if boolean_finding:      
            test = Mobile_data.loc[Mobile_data[0]== 'Memory Card Slot Type'].index[0]
            Memcardslot = Mobile_data.loc[test,1]
        else:
             Memcardslot = ''
    except AttributeError:
        Memcardslot = ''
#     print(Memcardslot)


    result = (Brand,original_price,discounted_price,discount,rating,rating_count,ModelName,ModelNumber,Simtype,Simsize,Technology,Touchscreen,Displaysize,PCAvailable,Primarycamera,SCAvailable,Secondarycamera,Intstorage,Ram,Exstorage,Memcardtype,Memcardslot,Battery)
    return result
    
    
    
```

```python


Brand = []
Original_price = []
Offered_Price = []
Discount = []
Rating = []
Rating_Count = []
Model_Name = []  
Model_Number = []
Sim_Type = []
Sim_Size = []
Tech_Category = []  # Samrtphone etc
Touch_Screen = []
Display_Size = []
PC_Available = []
Primary_Camera = []
SC_Available = []
Secondary_Camera = []
Internal_Storage = []
RAM = []
Expandable_Storage = []
Memory_Card_Type = []
Memory_Card_Slot = []
Battery = []

# No_of_Ratings = []
# No_of_Reviews = []







baseurl = 'https://www.flipkart.com'



for x in range(1,41):
    r = requests.get(f'https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&as-pos=1&as-type=RECENT&suggestionId=mobiles%7CMobiles&requestId=87868f22-b3aa-4db5-a84c-c4e55f2d39f3&as-searchtext=mobil&page={x}')
#     soup = BeautifulSoup(r.text,'html.parser')
    soup = BeautifulSoup(r.text,'html.parser')
    productlist = soup.find_all('div',attrs={'class':'_2kHMtA'})
#     print(len(productlist))
    for item in productlist:
        productlinks = []
        for link in item.find_all('a',href=True):
            productlinks.append(baseurl+link['href'])
        for url in productlinks:
            records = []
            record = extract_record(url)
#             print(record)
            if record:
                records.append(record)
                for info in records:
                    brand = info[0]
                    Brand.append(brand)
                    org_price = info[1]
                    Original_price.append(org_price)
                    off_price = info[2]
                    Offered_Price.append(off_price)
                    discount = info[3]
                    Discount.append(discount)
                    rating = info[4]
                    Rating.append(rating)
                    rating_count = info[5]
                    Rating_Count.append(rating_count)
                    ModelName = info[6]
                    Model_Name.append(ModelName)
                    ModelNumber = info[7]
                    Model_Number.append(ModelNumber)
                    Simtype = info[8]
                    Sim_Type.append(Simtype)
                    Simsize = info[9]
                    Sim_Size.append(Simsize)
                    Technology = info[10]
                    Tech_Category.append(Technology)
                    Touchscreen = info[11]
                    Touch_Screen.append(Touchscreen)
                    Displaysize = info[12]
                    Display_Size.append(Displaysize)
                    PCAvailable = info[13]
                    PC_Available.append(PCAvailable)
                    Primarycamera = info[14]
                    Primary_Camera.append(Primarycamera)
                    SCAvailable = info[15]
                    SC_Available.append(SCAvailable)
                    Secondarycamera = info[16]
                    Secondary_Camera.append(Secondarycamera)
                    Intstorage = info[17]
                    Internal_Storage.append(Intstorage)
                    Ram = info[18]
                    RAM.append(Ram)
                    Exstorage = info[19]
                    Expandable_Storage.append(Exstorage)
                    Memcardtype = info[20]
                    Memory_Card_Type.append(Memcardtype)
                    Memcardslot = info[21]
                    Memory_Card_Slot.append(Memcardslot)
                    battery = info[22]
                    Battery.append(battery)
            
                
dataset = {'Brand':Brand,'Model_Name':Model_Name,'Model_Number':Model_Number,'Original_Price':Original_price,
           'Offered_Price':Offered_Price,'Discount':Discount,'Rating':Rating,'Rating_Count':Rating_Count,
           'Sim_Size':Sim_Size,'Sim_Type':Sim_Type,'Tech_Category':Tech_Category,'Touch_Screen':Touch_Screen,
           'Display_Size':Display_Size,'PCam_Avl':PC_Available,'Primary_Camera':Primary_Camera,
           'SCam_avl':SC_Available,'Secondary_Camera':Secondary_Camera,'Int_Memory':Internal_Storage,
           'RAM':RAM,'Expandable_Memory':Expandable_Storage,'Memory_Card_Type':Memory_Card_Type,
           'Memory_Card_Slot':Memory_Card_Slot,'Battery':Battery}
data = pd.DataFrame(dataset)
print(len(data.index))
pd.set_option('display.max_colwidth',None)
print(data.head(1))
data.to_csv('Flipkart_Mobiles.csv')
```
